```python
import pandas as pd
import numpy as np
```

```python
MoodRasaDataFleksibel_df = pd.read_csv("MoodRasaDataFleksibel.csv")
MoodRasaDataFleksibel_df
```

```python
MoodRasaDataFleksibel_df.info()
```

```python
display(MoodRasaDataFleksibel_df.isnull().sum())
rows_with_nan = MoodRasaDataFleksibel_df[MoodRasaDataFleksibel_df.isnull().any(axis=1)]
display(rows_with_nan)
display(MoodRasaDataFleksibel_df.isnull())
```

```python
MoodRasaDataFleksibel_df['Rating'] = MoodRasaDataFleksibel_df['Rating'].fillna(MoodRasaDataFleksibel_df['Rating'].mean())
MoodRasaDataFleksibel_df['Waktu_Persiapan'] = MoodRasaDataFleksibel_df['Waktu_Persiapan'].fillna(MoodRasaDataFleksibel_df['Waktu_Persiapan'].mean())
MoodRasaDataFleksibel_df['Popularitas'] = MoodRasaDataFleksibel_df['Popularitas'].fillna(MoodRasaDataFleksibel_df['Popularitas'].mean())
MoodRasaDataFleksibel_df['Ketersediaan'] = MoodRasaDataFleksibel_df['Ketersediaan'].fillna(MoodRasaDataFleksibel_df['Ketersediaan'].mean())
MoodRasaDataFleksibel_df
```

```python
MoodRasaDataFleksibel_df.describe()
```

```python
display(MoodRasaDataFleksibel_df.isnull().sum())
rows_with_nan = MoodRasaDataFleksibel_df[MoodRasaDataFleksibel_df.isnull().any(axis=1)]
display(rows_with_nan)
```

```python
MoodRasaDataFleksibel_df.dropna(inplace=True)
display(MoodRasaDataFleksibel_df.isnull().sum())
```
